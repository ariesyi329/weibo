### Top Emotions Bar Plot

```python
kit_chen = utils.load_data(weibo_id=3794863604136940)
```

```python
index = kit_chen.index
text = kit_chen.text
emo_dict = utils.get_emotions(index, text)
emo_df = utils.dict_to_df(emo_dict, ['emo','freq'], 30)
```

```python
top = emo_df.head(10)
df_f = kit_chen.loc[kit_chen.user_gender == 'f']
df_m = kit_chen.loc[kit_chen.user_gender == 'm']
index_f = df_f.index
text_f = df_f.text
index_m = df_m.index
text_m = df_m.text
emo_dict_f = utils.get_emotions(index_f, text_f)
emo_dict_m = utils.get_emotions(index_m, text_m)
emo_df_f = utils.dict_to_df(emo_dict_f, ['emo','freq'], 30)
emo_df_m = utils.dict_to_df(emo_dict_m, ['emo','freq'], 30)
```

```python
li = []
for i in range(len(top)):
    for j in range(len(emo_df_f)):
        if emo_df_f['emo'][j] == top['emo'][i]:
            li.append(emo_df_f['freq'][j]/len(df_f))
            break
top['f'] = li

li2 = []
for i in range(len(top)):
    for j in range(len(emo_df_m)):
        if emo_df_m['emo'][j] == top['emo'][i]:
            li2.append(emo_df_m['freq'][j]/len(df_m))
            break
top['m'] = li2
```

```python
top['ratio'] = top['f']/top['m']
top['en'] = ['love u','heart','good','cry','like','flowers','applause','awesome','shy','hehe']
top = top.set_index('en')
top = top.sort(columns='freq',ascending=True)
```

```python
bar('freq', top, title='''Emotions from Kit Chan's Weibo''', xlabel='frequency')
plt.savefig('../results/emotions.png', bbox_inches='tight', pad_inches = 0.5, dpi=300)
```

### User weibo behavior vs. user weibo relationships

```python
boxprep = kit_chen[['user_favourites_count','user_followers_count','user_friends_count','user_statuses_count','user_gender']]
```

```python
follower_group = pd.Series()
for i in boxprep.index:
    if boxprep.ix[i].user_followers_count <= boxprep.describe().user_followers_count['25%']:
        group = 1
    elif boxprep.ix[i].user_followers_count <= boxprep.describe().user_followers_count['50%']:
        group = 2
    elif boxprep.ix[i].user_followers_count <= boxprep.describe().user_followers_count['75%']:
        group = 3
    else:
        group = 4
    follower_group = follower_group.set_value(i,group)
```

```python
friends_group = pd.Series()
for i in boxprep.index:
    if boxprep.ix[i].user_friends_count <= boxprep.describe().user_friends_count['25%']:
        group = 1
    elif boxprep.ix[i].user_friends_count <= boxprep.describe().user_friends_count['50%']:
        group = 2
    elif boxprep.ix[i].user_friends_count <= boxprep.describe().user_friends_count['75%']:
        group = 3
    else:
        group = 4
    friends_group = friends_group.set_value(i,group)
```

```python
boxprep['user_follower_group'] = follower_group
boxprep['user_friends_group'] = friends_group
boxprep2 = boxprep.loc[(boxprep.user_statuses_count < 4000)]
boxprep3 = boxprep.loc[(boxprep.user_favourites_count < 600)]
```

```python
g = factorplot(x = "user_friends_group", y = "user_favourites_count", hue = "user_gender", data = boxprep3,
               xlabel = 'Groups in Friends Count', ylabel = 'Favourites Count', legend_title = 'Gender')
groups = ["< 113", "113 ~ 211", "211 ~ 343", "> 343"]
g.set_xticklabels(groups)
plt.title('Favourites Count vs. Friends Count', fontsize = 16, y=1.05)
plt.savefig('../results/dist2.png', bbox_inches='tight', pad_inches = 0.5, dpi=300)
```

```python
g = factorplot(x = "user_follower_group", y = "user_statuses_count", hue = "user_gender", data = boxprep2,
               xlabel = 'Groups in Followers Count', ylabel = 'Statuses Count', legend_title = 'Gender')
groups = ["< 82", "82 ~ 230", "230 ~ 496", "> 496"]
g.set_xticklabels(groups)
plt.title('Statuses Count vs. Followers Count', fontsize = 16, y=1.05)
plt.savefig('../results/dist3.png', bbox_inches='tight', pad_inches = 0.5, dpi=300)
```

### Comments Trend

```python
plotprep = kit_chen[['created','user_gender']]
```

```python
plotprep['time'] = plotprep['created'].apply(lambda i: datetime.datetime.strptime(i, "%a %b %d %H:%M:%S +0800 %Y"))
plotprep['counts'] = np.ones(len(plotprep))
time = plotprep.groupby(plotprep['time']).sum()
time = time.groupby((time.index.day, time.index.hour)).sum()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
time['counts'].plot(kind='area',alpha=0.6,legend=False,linewidth=1.5,color='Skyblue')
utility.axis_helper(ax, color='dimgrey')
labels = ['16, Jan 3', '00, Jan 5', '22, Jan 5', '23, Jan 6', '23, Jan 7', '14, Jan 9', '17, Jan 10']
ax.set_xticklabels(labels)
ax.xaxis.label.set_color('dimgrey')
ax.yaxis.label.set_color('dimgrey')
utility.title_helper(ax, '''Comments Trend at Kit Chan's Weibo''', ha='left', xloc=0)
plt.annotate(
    '''AMOST ONE DAY AFTER\nKIT'S FIRST SHOW AT\nTHE STAGE OF I AM A SINGER''',
    xy=(2, max(time['counts'])), arrowprops=dict(arrowstyle='->'), xytext=(10, max(time['counts'])+3), fontsize=10)
plt.annotate(
    '''KIT OUT OF THE CONTEST''',
    xy=(108, 106), arrowprops=dict(arrowstyle='->'), xytext=(105, 115), fontsize=10)
plt.savefig('../results/trend.png', bbox_inches='tight', pad_inches = 0.5, dpi=300)
```

### User distribution : pie chart

```python
pieprep = kit_chen[['user_created','user_gender']]
```

```python
pieprep['time'] = pieprep['user_created'].apply(lambda i: datetime.datetime.strptime(i, "%a %b %d %H:%M:%S +0800 %Y"))
pieprep['count'] = np.ones(len(pieprep))
pieprep = pieprep.set_index('time')
user_time = pieprep.groupby(pieprep.index.year).sum()
```

```python
user_time = user_time.drop(user_time.index[6])
user_time.columns = ['']
```

```python
plt.xkcd()
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111)
plt.title('User Distribution: Account Creation Year', fontsize=20)
colors = ["#fbb4ae",
"#b3cde3",
"#ccebc5",
"#decbe4",
"#fed9a6",
"#ffffcc"]
explode = (0, 0.1, 0, 0,0,0)
user_time[''].plot(kind='pie', colors=colors, fontsize=14, explode=explode,autopct='%1.1f%%', shadow=False, startangle=90)
plt.savefig('../results/pie1.png', bbox_inches='tight', pad_inches = 0.5, dpi=300)
```
